```python
def tokenize_line(line):

    def parse_coord(coord):
        start, end = coord[2:].split("..")
        return int(start), int(end)

    on_off, coords = line.split(" ")
    x, y, z = coords.split(",")
    return on_off, parse_coord(x), parse_coord(y), parse_coord(z)

def parse_text(raw_text):
    return [tokenize_line(l) for l in raw_text.split("\n") if l]

def read_input():
    with open("input.txt", "rt") as f:
        return f.read()

def split_input(lines):
    return lines[0], lines[1:]
```

```python
def range_inclusive(start, end):
    return range(start, end + 1)

def range_is_valid(start, end):
    return abs(start) <= 50 and abs(end) <= 50

def ranges_are_valid(*ranges):
    return all(range_is_valid(*range) for range in ranges)

def mark_points(lines):
    cube = defaultdict(int)
    for on_off, x_range, y_range, z_range in lines:
        if not ranges_are_valid(x_range, y_range, z_range):
            print("Skipping", on_off, x_range, y_range, z_range)
            continue
        # print(on_off, x_range, y_range, z_range)
        if on_off == 'on':
            setting = 1
        else:
            setting = 0
        for x, y, z in product(range_inclusive(*x_range), range_inclusive(*y_range), range_inclusive(*z_range)):
            coord = x, y, z
            # print(setting, coord)
            if setting:
                cube[coord] = setting
            else:
                if coord in cube:
                    del cube[(x, y, z)]
    return cube
```

```python
# part 2
# inspired by github.com/topaz

def count(cubes):
    if not cubes:
        return 0

    (state, *head), *tail = cubes
    # Don't return the volume of any "off" cubes
    if state == 'off':
        return count(tail)

    # For "on" cubes, return the volume less the intersection with other cubes
    # Since we're not including the volume of "off" cubes, they won't be included
    # in the final result
    return volume(*head) + count(tail) - count(
       {intersect(*head, *t) for t in tail}-{None})

def is_positive_volume(x2, y2, z2):
    return x2[0] <= x2[1] and y2[0] <= y2[1] and z2[0] <= z2[1]

def intersect(x2, y2, z2, _, ox2, oy2, oz2):
    nx2 = (
        max(x2[0], ox2[0]),
        min(x2[1], ox2[1])
    )
    ny2 = (
        max(y2[0], oy2[0]),
        min(y2[1], oy2[1])
    )
    nz2 = (
        max(z2[0], oz2[0]),
        min(z2[1], oz2[1])
    )

    if is_positive_volume(nx2, ny2, nz2):
        return '', nx2, ny2, nz2

def volume(x2, y2, z2):
    return (x2[1] - x2[0] + 1) * (y2[1] - y2[0] + 1) * (z2[1] - z2[0] + 1)

count(parse_text(SAMPLE_TEXT_BIG))
```
